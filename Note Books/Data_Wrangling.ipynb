<h2>Exploring missingness:</h2>

```python
#checking missingness by column 
def percentageOfnull(df):
    nan_count = df.isna().sum().sort_values(ascending=False)
    nan_percentage = ((df.isna().sum()/df.isna().count())*100).sort_values(ascending=False)
    train_test_dtype = df.dtypes
    df_info = pd.concat([nan_count,nan_percentage,train_test_dtype],axis=1,keys=['Null Count','Null Percentage %','DataTypes'])
    return df_info[df_info['Null Count']>0]
```

```python
percentageOfnull(data)
```

By examining the column discriptions, we can see that 'PoolQC' and the rest of the catagorical data aren't missing but rather aren't available features for the houses. In other words 96% of 'PoolQC' is missing because 96% of houses don't have pools.

<h2>Data Cleaning:</h2>

Since all catogerical values aren't really missing, but rather not included features in the houses, we will fill all thir missing values with 'none' to indicate this. 

```python
#We will select all categorical columns and replace missing values with 'none':
cat_data = list(data.select_dtypes(exclude='number'))
print(cat_data)
num_data = list(data.select_dtypes(include='number'))
print(num_data)
```

```python
data[cat_data] = data[cat_data].replace(np.nan, 'none')
data[cat_data].head()
```

```python
percentageOfnull(data[cat_data])
```

Now we will examine and fill in the numerical values:

```python
percentageOfnull(data[num_data])
```

```python
%matplotlib inline
data[['LotFrontage','GarageYrBlt','MasVnrArea']].hist(bins=50, figsize=(20,20))
plt.show()
```

```python
data[['LotFrontage','GarageYrBlt','MasVnrArea']].describe()
```

'LotFrontage', 'MasVnrArea' both have outliers, and 'GarageYrBlt' is skewed to the left.
With this in mind it seems that the median is the best option to replace missing values for all three. 

```python
#the medians are: 
print(data['LotFrontage'].median())
print(data['GarageYrBlt'].median())
print(data['MasVnrArea'].median())
```

```python
#fillinf in the median:
for i in ['LotFrontage','GarageYrBlt','MasVnrArea']:
    data[i].fillna(data[i].median(), inplace=True)
```
